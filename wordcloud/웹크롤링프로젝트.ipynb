```python
ind = 1  # 현재 복사한 순서
no = 1  # 1~10페이지 중 위치한 곳
page = 1  # 현재 페이지 번호
list1  = []  # 결과물이 저장되는 리스트
list2 = []

             # title_click : 글목록으로 들어가기 
             # title : 제목 
             # training_name : 연수명
             # Classification : 분류
             # rank : 만족도
             # content : 본문내용
             # back : 목록으로 돌아가기 
while page <= 300:                
                                                            
        try:
            title_click = driver.find_element(by='xpath', value = f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[3]/a').click()
            title = driver.find_element(by='xpath',value = f'//*[@id="bbs_wrap"]/div/dl[1]/dd').text
            training_name = driver.find_element(by ='xpath', value = f'//*[@id="bbs_wrap"]/div/dl[2]/dd/a').text
            Classification= driver.find_element(by ='xpath', value = f'//*[@id="bbs_wrap"]/div/dl[3]/dd[2]').text
            rank_element = driver.find_element(By.CSS_SELECTOR, "#bbs_wrap > div > dl:nth-child(3) > dd:nth-child(2) > span > span")
            rank = rank_element.get_attribute('class').split('-')[-1][-1]
            content = driver.find_element(by ='xpath', value = f'//*[@id="bbs_wrap"]/div/div[1]').text
            back = driver.find_element(by='xpath', value = f'//*[@id="btn-list"]').click()
            list2.append([title, training_name, Classification,rank,content])
                                                    
            ind += 1
            
                        
        except:
            # (페이지 이동)
            try:
                no += 1
                xpath_next_page = f'//*[@id="paging"]/ul/li[{no+1}]/a'
                driver.find_element(by='xpath', value=xpath_next_page).is_displayed()
                driver.find_element(by='xpath', value=xpath_next_page).click()
                ind = 1
                page += 1
                
                continue
                
            except:               # 페이지 번호가 10이넘어가면 다음화살표를 클릭 
                if no >= 10:
                    xpath_last_page = '//*[@id="paging"]/ul/li[12]/a'
                    driver.find_element(by='xpath', value=xpath_last_page).click()
                    no = 1
                    ind = 1
                    page += 1                   
                    
                    continue

            break
```
